```python
%%writefile funcion.cpp
#include <iostream>
using namespace std;

int funcion1(int n){ //tiempo constante
  return n*(n+1)/2;
}

int funcion2(int n){//tiempo lineal
  int s=0;
  for(int i=1;i<=n;++i) s+=i;
  return s;  
}

int funcion3(int n){//tiempo cuadratico
  int s=0;
  for(int i=1;i<=n;i++){
      for(int j=0;j<i;j++){
          s+=1;
      }
  } 
  return s;
}
int main(){   
  cout << funcion1(3)<<endl;  
  cout << funcion2(3)<<endl;  
  cout << funcion3(3);  
  return 0;
}

```
